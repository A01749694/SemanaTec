```python
numeric_columns = []
# for col in df.columns:
#     if df[col].dtype != "O":
#         numeric_columns.append(col)
numeric_columns = [col for col in df.columns if df[col].dtype != "O"] #Crear un arreglo con todos los elemento de la columna con list comprehension
object_columns = [col for col in df.columns if df[col].dtype == "O"]
```

```python
df[numeric_columns].plot.box()
```

```python
for col in numeric_columns: 
    df[col].plot.box() #Plotea la gráfia de bigote con los datos de las columnas númericas
    plt.show()
```

```python
for col in numeric_columns:
    df[col].plot.kde()
    plt.show() #Graficar la densidad de los datos 
```

```python
print("Medidas de tendencia central")
print(f"Suma : {df["Year"].sum()}")
print(f"Promedio : {df["Year"].mean()}")
print(f"Mediana : {df["Year"].median()}")
print(f"Moda : {df["Year"].mode()}") #No se acostumbra con datos númericos
print("Medidas de dispersión")
print(f"Desviación estándar : {df["Year"].std()}")
print(f"Varianza : {df["Year"].var()}")
print("Medidas de posición")
print(f"min: {df["Year"].min()}")
print(f"Q1: {df["Year"].quantile(0.25)}")
print(f"Q2: {df["Year"].quantile(0.5)}")
print(f"Q3: {df["Year"].quantile(0.75)}")
print(f"max: {df["Year"].max()}")
```

```python
#df[máscara] para consultas
df[df["Year"] <= 1990][["Name", "Platform"]]
```

```python
#todos los juegos publicados por nintendo 
# / -> or
# & -> and
# ~ -> not
df[(df["Year"] > 2000) & (df["Publisher"] == "Nintendo")]["Name"]
```

```python
#Mostrar todos los juegos del ps1/psx, saturn y n64
#df["Platform"].unique() #Muestra todos los valores de la columna sin repetición
# df[(df["Platform"] == "PS") | (df["Platform"] == "N64") | (df["Platform"] == "SAT")]["Name"]
df[df["Platform"].isin(["PS", "N64", "SAT"])]["Name"]
```

```python
df["NA_Sales"].nlargest(3) # n datos más grandes
df["NA_Sales"].nsmallest(5) #n datos más pequeños
df.nsmallest(n=5, columns=["NA_Sales"])
```
